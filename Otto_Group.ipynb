## Data Preparation


```
from sklearn.preprocessing import StandardScaler
```

```
scaler = StandardScaler()
```

```
scaler.fit(df.drop(['id','target'],axis=1))
```

```
scaled_features = scaler.fit_transform(df.drop(['id','target'],axis=1))
```

```
df_feat = pd.DataFrame(scaled_features,columns=df.columns[1:-1])
df_feat.head()
```

## Training Test Split

```
X = df_feat
```

```
y = df['target']
```

```
X = X.as_matrix()
y = y.as_matrix()
```

```
from sklearn.model_selection import train_test_split
```

```
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

## Contrib.learn

```
import tensorflow.contrib.learn as learn
```

```
from sklearn import model_selection
from sklearn import metrics
import tensorflow as tf
```

```
feature_columns = tf.contrib.learn.infer_real_valued_columns_from_input(X_train)

classifier = tf.contrib.learn.DNNClassifier(feature_columns=feature_columns, hidden_units=[50, 50, 50], n_classes=9)
```

```
# Fit and predict.
classifier.fit(X_train, y_train, steps=1000, batch_size=20)
```

```
predictions = np.array(list(classifier.predict(X_test, as_iterable=True)))
score = metrics.accuracy_score(y_test, predictions)
print('Accuracy: {0:f}'.format(score))
```

```
pred = list(classifier.predict_proba(X_test))
test_loss = metrics.log_loss(y_test,pred) 
test_loss
```

```
from sklearn.metrics import classification_report, confusion_matrix
```

```
print(confusion_matrix(y_test, predictions))
```

```
print(classification_report(y_test, predictions))
```
