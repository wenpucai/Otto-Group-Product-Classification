```
# main libraries
import pandas as pd
import numpy as np
# sklearn libraries
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```
import requests
import io

response = requests.get('http://www.utdallas.edu/~jxz172130/data/train.csv')

file_object = io.StringIO(response.content.decode('utf-8'))
df = pd.read_csv(file_object)
```

```
df.head()
```

### converting categorical to numerical values


```
cleanup_nums = {"target": {"Class_1": 0, "Class_2": 1, "Class_3": 2, "Class_4": 3, "Class_5": 4, "Class_6": 5, "Class_7": 6, "Class_8": 7, "Class_9": 8}}
df.replace(cleanup_nums, inplace=True)
df.head()
```

# Data Preparation

```
from sklearn.preprocessing import StandardScaler
```

```
scaler = StandardScaler()
```

```
scaler.fit(df.drop(['id','target'],axis=1))
```

```

scaled_features = scaler.fit_transform(df.drop(['id','target'],axis=1))
```

```
df_feat = pd.DataFrame(scaled_features,columns=df.columns[1:-1])
df_feat.head()
```

```
# nrows, ncols = df.shape[0], df.shape[1]
# X = df.iloc[:, 1:(ncols - 1)].values.reshape(nrows, ncols - 2)
# y = df.iloc[:, (ncols - 1)].values.reshape(nrows, 1)
```

```
# X
```

# Training Test Split

```
X = df_feat
y = df['target']
```

```
X = X.as_matrix()
y = y.as_matrix()
```

```
from sklearn.model_selection import train_test_split
```

```
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

# Train Dataset

```
from xgboost import XGBClassifier
```

```
classifier = XGBClassifier(base_score=0.5, 
                           booster='gbtree', 
                           colsample_bylevel=1,
                           colsample_bytree=1, 
                           gamma=0, 
                           learning_rate=0.3, 
                           max_delta_step=0,      
                           max_depth=4, 
                           min_child_weight=1, 
                           missing=None, 
                           n_estimators=100,     
                           n_jobs=1, 
                           nthread=None, 
                           objective='multi:softprob', 
                           random_state=0,      
                           reg_alpha=0, 
                           reg_lambda=1, 
                           scale_pos_weight=1, 
                           seed=None,
                           silent=False, 
                           subsample=1)
```

```
classifier.fit(X_train, y_train, eval_metric='mlogloss', verbose=True,
            eval_set=[(X_test, y_test)], early_stopping_rounds=100)
```

# Prediction of Test Set

```
import requests
import io

response = requests.get('http://www.utdallas.edu/~jxz172130/data/test.csv')

file_object = io.StringIO(response.content.decode('utf-8'))
dfTest = pd.read_csv(file_object)

dfTest = dfTest.drop('id',axis=1)
```

```
y_pred = classifier.predict(X_test)
```

```
from sklearn import metrics
```

```
score = metrics.accuracy_score(y_test, y_pred)
print('Accuracy: {0:f}'.format(score))
```

```
from sklearn.metrics import classification_report, confusion_matrix
```

```
print(confusion_matrix(y_test, y_pred))
```

```
print(classification_report(y_test, y_pred))
```
